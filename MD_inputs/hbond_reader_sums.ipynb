```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
frame_count = 279

def read_hbonds(exp_id, frame_count):
    hbonds = [np.loadtxt(exp_id+'/'+exp_id+"_frame"+str(frame)+"_hbonds.dat")[:,1] for frame in range(frame_count)]
    return np.vstack(hbonds).T
```

```python
for exp_count in range(1,46):
    if exp_count == 1:
        hbonds = read_hbonds(f'oxa{exp_count}',frame_count)
    else:
        hbonds += read_hbonds(f'oxa{exp_count}',frame_count)
hbonds
#np.savetxt("hbonds.dat",hbonds)
```

```python
residual_sums = hbonds.sum(axis=1)

avg_res_sum = residual_sums/45

nonzero_residues = np.where(residual_sums>0)[0]
nonzero_avg_res_sum = np.where(avg_res_sum>0)[0]
nonzero_residues


np.savetxt("oxa_residtotal_hbonds.dat",residual_sums)
np.savetxt("oxa_avg_resid_hbonds.dat",avg_res_sum)
```
